```python
import numpy as np
%pylab inline
```

```python
T = 1
N = 200
M = 2 # number of paths
t = np.linspace(0,T,N+1)
dt = np.diff(t).reshape((N,1))
dt = np.tile(dt,(1,M))
```

```python
mu = 1
sigma = 0.5
alpha = 5.4568386739546
X0 = 2
ealph = exp(-alpha*dt)
ealph2 = exp(-2*alpha*dt)
```

```python
X = np.ones((N+1,M))*X0
for n in range(N):
    X[n+1,:] = X[n,:]*ealph[n,:] + mu*(1-ealph[n,:])
    X[n+1,:] += (sigma/np.sqrt(2*alpha))*np.sqrt(1-ealph2[n,:])*\
    np.random.randn(1,M).reshape(M,)
plot(t,X);
```
